```python
import requests

# 1. PASTE YOUR REAL SECTORS.APP KEY HERE
MY_API_KEY = "da2d9b658bb0491364447738c93bb3535eea81ae2cf3e26e4711331df8c1746a"

# 2. Add the "Bearer " prefix
headers = {"Authorization": f"Bearer {MY_API_KEY}"}

url = "https://api.sectors.app/v1/company/BMRI.JK/profile"

print(f"Testing API call to: {url}")
# Hides most of your key
print(f"Using header: Authorization: Bearer {MY_API_KEY[-4:]}")

try:
    response = requests.get(url, headers=headers)
    response.raise_for_status()  # This will raise an error if it's 4xx or 5xx

    data = response.json()
    print("\n--- SUCCESS! ---")
    print("Successfully connected and fetched data.")
    print(f"Company Name: {data.get('data', {}).get('name')}")

except requests.exceptions.HTTPError as err:
    print("\n--- ERROR! ---")
    if err.response.status_code == 401:
        print("HTTP 401: Unauthorized. This confirms the API key is invalid or has not been activated.")
    elif err.response.status_code == 404:
        print("HTTP 404: Not Found. This means the URL is wrong (unlikely) or the 'Bearer' prefix was missing.")
    else:
        print(f"An HTTP error occurred: {err}")
except Exception as e:
    print(f"A non-HTTP error occurred: {e}")
```

```python
import os
import json
from dotenv import load_dotenv
from newsapi import NewsApiClient

# --- 1. Load Environment Variables ---
load_dotenv()
NEWS_API_KEY = os.getenv("NEWS_API_KEY")

# --- 2. Check if the API Key was found ---
if not NEWS_API_KEY:
    print("FATAL ERROR: NEWS_API_KEY not found in your .env file.")
    print("Please ensure the key is present and correct.")
    exit()

print(f"Found API Key, starting with prefix: {NEWS_API_KEY[:4]}...")

# --- 3. Initialize the Client and Make the API Call ---
try:
    # Initialize the client
    newsapi = NewsApiClient(api_key=NEWS_API_KEY)

    print("\nFetching news for 'IDX Finance'...")

    # Get top headlines related to the company name
    headlines = newsapi.get_everything(
        q='IDX Finance',
        language='en',
        sort_by='relevancy',
        page_size=3
    )

    # --- 4. Print the Full Response ---
    print("\n--- ✅ SUCCESS: Received Response from NewsAPI ---")
    # Use json.dumps for pretty-printing the dictionary
    print(json.dumps(headlines, indent=2))


except Exception as e:
    print("\n--- ❌ ERROR: An exception occurred ---")
    print(f"The API call failed. The error was: {e}")
    print("\nPlease double-check that your NEWS_API_KEY in the .env file is correct, active, and has no extra spaces or characters.")
```
